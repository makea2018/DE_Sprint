# Импорт библиотек

```python
import numpy as np
import matplotlib.pyplot as plt
```

# Переменные

```python
x = np.arange(-20, 20, 0.001)
y_1 = 0.5 * ((x**2)**(1/3) + ((x**4)**(1/3) - 4 * x**2 + 4)**0.5)
y_2 = 0.5 * ((x**2)**(1/3) - ((x**4)**(1/3) - 4 * x**2 + 4)**0.5)
```

# Строим график

```python
plt.plot(x, y_1, color = "red")
plt.plot(x, y_2, color = "red")
plt.show()
```
