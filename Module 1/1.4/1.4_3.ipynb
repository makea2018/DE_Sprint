# Импорт библиотек

```python
import numpy as np
import matplotlib.pyplot as plt
```

# Списки данных

```python
age = []
sex = []
bmi = []
children = []
smoker = []
whereabouts = []
insurance = []
```

# Считываем данные и добавляем их в списки по категориям

```python
with open ("D:/Курсы по программированию 2022г/Инженер данных 2022/DE_Sprint/Module 1/1.4/dataset_home.txt", mode = "r") as file:
    text_arr = file.readlines()
    for human in range(0, len(text_arr)):
        age.append(int(text_arr[human].split(',')[0]))
        if text_arr[human].split(',')[1] == 'male':
            sex.append(1)
        else:
            sex.append(0)
        bmi.append(float(text_arr[human].split(',')[2]))
        children.append(int(text_arr[human].split(',')[3]))
        if text_arr[human].split(',')[4] == 'yes':
            smoker.append(1)
        else:
            smoker.append(0)    
        if text_arr[human].split(',')[5] == 'northwest':
            whereabouts.append(0)
        elif text_arr[human].split(',')[5] == 'northeast':
            whereabouts.append(1)
        elif text_arr[human].split(',')[5] == 'southwest':
            whereabouts.append(2) 
        elif text_arr[human].split(',')[5] == 'southeast':
            whereabouts.append(3)
        insurance.append(float(text_arr[human].split(',')[6]))
```

# Список превращаем в массив

```python
age = np.array(age)
sex = np.array(sex)
bmi = np.array(bmi)
children = np.array(children)
smoker = np.array(smoker)
whereabouts = np.array(whereabouts)
insurance = np.array(insurance)
```

# График зависимости кол-ва детей от возраста человека

```python
plt.scatter(age[children <= 1], children[children <= 1], color="blue", marker='o', label = " children < 1")
plt.scatter(age[children > 1][children[children > 1] <= 3], children[children > 1][children[children > 1] <= 3], color="yellow", marker='o', label = "1 <= children <= 3")
plt.scatter(age[children > 3][children[children > 3] <= 6], children[children > 3][children[children > 3] <= 6], color="red", marker='o', label = "3 <= children <= 6")
plt.title("Кол-во детей у людей разного возраста")
plt.xlabel("Возраст")
plt.ylabel("Кол-во детей")
plt.legend(bbox_to_anchor=( 1.02 , 0.18 ), loc='upper left', borderaxespad= 0 )
plt.show()
```

# График зависимости индекса массы тела от возраста

```python
plt.scatter(age[bmi < 20], bmi[bmi < 20], color = "green", label = "bmi < 20")
plt.scatter(age[bmi >= 30][bmi[bmi >= 30] <= 40], bmi[bmi >= 30][bmi[bmi >= 30] <= 40], color= "pink", label = "30 <= bmi <= 40")
plt.title("Зависимость индекса массы тела от возраста")
plt.xlabel("Возраст")
plt.ylabel("BMI")
plt.legend(bbox_to_anchor=( 1.02 , 0.12 ), loc='upper left', borderaxespad= 0 )
plt.show()
```

# График зависимости страховки от возраста

```python
plt.scatter(np.log(insurance[age >= 20][age[age >= 20] <= 60]), age[age >= 20][age[age >= 20] <= 60], color = "skyblue", label = "20 <= age < 60")
plt.title("Зависимость размера страховки от возраста")
plt.xlabel("Страховка, $")
plt.ylabel("Возраст, лет")
plt.legend(bbox_to_anchor=( 1.02 , 0.07 ), loc='upper left', borderaxespad= 0 )
plt.show()
```

# Столбчатая диаграмма, какая страховка у курящих и некурящих

```python
plt.bar(smoker, insurance, width= 0.6)
plt.xlabel("Курят или нет")
plt.ylabel("Размер страховки, $")
plt.show()
```
