# Импорт библиотек

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Носитель или выборка
x = np.arange(0, 5, 0.05)

# Плотность или масштаб
G = np.arange(0.5, 4.1, 0.5)
```

# Функция распределения плотности

```python
for i in G:
    y = 1 - np.exp(-x**2/2*i**2)
    plt.plot(x,y, label = f"G = {i}")
    plt.legend()
    plt.show
```

# Плотность вероятности

```python
for i in G:
    y = np.random.rayleigh(i, 40000)
    plt.figure()
    plt.hist(y, bins = 100, density = True, label= f"G = {i}")
    plt.legend()
    plt.show()
```
