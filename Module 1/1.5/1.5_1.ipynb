# **Cредняя и медианная зарплата по каждому региону**

```python
States_list = df["company_location"].unique()
States_list
```

```python
for i in States_list:
  # Средняя зарплата в каждом штате
  state = i
  mean_sal = df[df["company_location"] == state].salary.mean()
  median_sal = df[df["company_location"] == state].salary.median()
  print(f"Средняя зарплата в регионе {i} = {mean_sal} тыс. евро\nМедианная зарплата в регионе {i} = {median_sal} тыс. евро\n")
  print("--------------------------------------------------------------------------")
```

# **Какая самая высокооплачиваемая из групп вакансий, исходя из их средних зарплат?**

```python
# Средние зарплаты для анализа
mean_sal_Data_Scientist = Data_scientist.salary.mean()
mean_sal_Data_Engineer = Data_engineer.salary.mean()
mean_sal_Data_Analyst = Data_analyst.salary.mean()
mean_sal_dict = {"Data Scientist": mean_sal_Data_Scientist, "Data Engineer": mean_sal_Data_Engineer, "Data Analyst": mean_sal_Data_Analyst}
vacancyMaxSalary = max(mean_sal_dict, key = mean_sal_dict.get)
print("Максимальная зарплата, исходя из средних зарплат, у", vacancyMaxSalary)
```

# **Какое процентное соотношение каждого региона по вакансиям от всех вакансий?**

```python
# Кол-во вакансий
Vac_all_quantity = df.salary.count()

# Проходим по каждому региону

for i in States_list:
  
  # Кол-во вакансий в каждом штате
  state = i
  quantity = df[df["company_location"] == state].salary.count()
  print(f"Кол-во вакансий в регионе {state} =  {quantity}" )
  
  # Находим процентное отношение вакансий в регионе относительно всех вакансий
  ratio = float(quantity) / 100
  print(f"Процентное соотношение вакансий региона {state} относительно всех вакансий:  {ratio}%")
  print("--------------------------------------------------------------------------")
```

# **Какая корреляция уровня опыта от зарплаты?**

```python
df_Salary_Expreience = pd.DataFrame(df, columns = ["experience_level", "salary"])

# Узнаем средние зарплаты, чтобы примерно понять, как перевести вакансии по опыту из аббревиатур в цифры
d = df_Salary_Expreience[df_Salary_Expreience["experience_level"] == 'EX'].salary.median() # EX
a = df_Salary_Expreience[df_Salary_Expreience["experience_level"] == 'SE'].salary.median() # SE
b = df_Salary_Expreience[df_Salary_Expreience["experience_level"] == 'MI'].salary.median() # MI
c = df_Salary_Expreience[df_Salary_Expreience["experience_level"] == 'EN'].salary.median() # EN


print("{} - средняя зарплата EX\n{} - средняя зарплата SE\n{} - средняя зарплата MI\n{} - средняя зарплата EN\n".format(d, a, b, c))
# Преобразуем experience в цифры []
```

```python
# Меняем текстовые значения experience на числовые
''' EX = 5 лет
    SE = 4 года
    MI = 2 года
    EN = 1 год    
'''
df_Salary_Expreience.loc[df_Salary_Expreience["experience_level"] == "EX", "experience_level"] = 5
df_Salary_Expreience.loc[df_Salary_Expreience["experience_level"] == "SE", "experience_level"] = 4
df_Salary_Expreience.loc[df_Salary_Expreience["experience_level"] == "MI", "experience_level"] = 2
df_Salary_Expreience.loc[df_Salary_Expreience["experience_level"] == "EN", "experience_level"] = 1

# Находим корреляцию опыта от зарплаты
df_Salary_Expreience['salary'].astype('float64').corr(df_Salary_Expreience['experience_level'].astype('float64'))
```
